```python
# os.listdir('./results/')
reinforce = [os.path.join('./results/', f) for f in [
    '201912152113-gamma-0.5-lr-0.05-nodropout.json',
    '201912151824-IN24232-OUT623-gamma-0.5-lr-0.01.json',
    '201912152249-gamma-0.5-lr-0.01-nodropout.json',
    '201912161346-gamma-0.5-lr-0.001-nodropout.json',
    '201912161049-gamma-0.5-lr-0.01-nodropout.json',
    '201912151724-IN24232-OUT623-gamma-0.5-lr-0.01.json',
    '201912151840-gamma-0.5-lr-0.01-nodropout.json',
    '201912152119-gamma-0.5-lr-0.05-nodropout.json',
    '201912151721-IN24232-OUT623-gamma-0.5-lr-0.01.json',
#     '201912151720-IN24232-OUT623-gamma-0.5-lr-0.01.json',
    '201912151758-IN24232-OUT623-gamma-0.5-lr-0.01.json',
    '201912151723-IN24232-OUT623-gamma-0.5-lr-0.01.json'
]]

def read_reinforce(fname):
    print(fname)
    with open(fname) as f:
        d = json.load(f)
        df = pd.DataFrame({'rewards': d['rewards'], 'fname': fname})
        return df

df = pd.concat([read_reinforce(f) for f in reinforce])
```

```python
with open('./results/201912161346-gamma-0.5-lr-0.001-nodropout.json') as f:
    d = json.load(f)
df = pd.DataFrame({'rewards':d['rewards']})
plt.title("REINFORCE moving average reward")
df.rewards.rolling(window=100, win_type='triang').mean().plot();
plt.ylim(-1.05, -0.5)

wins = len(d['wins'])
games = len(d['rewards'])
print(f"{wins} wins, {games} games, {wins*100/games:.1f}% win rate")
```
