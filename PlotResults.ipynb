```python
import json
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
with open('./results/201912152249-gamma-0.5-lr-0.01-nodropout.json') as f:
    d = json.load(f)
df = pd.DataFrame({'rewards':d['rewards']})
plt.title("REINFORCE moving average reward")
df.rewards.rolling(window=100, win_type='triang').mean().plot();

wins = len(d['wins'])
games = len(d['rewards'])
print(f"{wins} wins, {games} games, {wins*100/games:.1f}% win rate")
```
