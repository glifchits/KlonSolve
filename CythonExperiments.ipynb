```python
%%cython --annotate
DEF SUIT = 1

cdef int can_stack(str card, str onto):
    """ `card` is the card to move, `onto` is the card to stack onto """
    print('card', card, 'onto', onto)
    cdef char cs = card[SUIT]
    cdef char os = onto[SUIT]
    print('cs', cs, 'os', os)
    cdef int C = 67
    cdef int D = 68
    cdef int S = 83
    cdef int H = 72
    if (cs == D or cs == H) and (os == D or os == H):
        return 0
    if (cs == C or cs == S) and (os == C or os == S):
        return 0
    return 1

c1 = u'AD'
c2 = u'JH'
print(can_stack(c1, c2))
```

```python
%%cython

cdef enum Suit:
    C=0, D, S, H

# values: A=1, 2=2, 3=3, ... 10=10, J=11, Q=12, K=13

cdef class Card:
    cdef Suit suit
    cdef int value
    cdef int faceup
    
    def __init__(self, card_string):
        value, suit = card_string
        # suit
        if suit.upper() == "C":
            self.suit = Suit.C
            self.faceup = suit == "C"
        elif suit.upper() == "D":
            self.suit = Suit.D
            self.faceup = suit == "D"
        elif suit.upper() == "S":
            self.suit = Suit.S
            self.faceup = suit == "S"
        elif suit.upper() == "H":
            self.suit = Suit.H
            self.faceup = suit == "H"
        # val
        if value.upper() == "A": 
            self.value = 1
        elif value.upper() == "T":
            self.value = 10
        elif value.upper() == "J":
            self.value = 11
        elif value.upper() == "Q":
            self.value = 12
        elif value.upper() == "K":
            self.value = 13
        else:
            self.value = int(value)
    
    def get_suit(self):
        if self.suit == Suit.C:
            suit_str = "C"
        elif self.suit == Suit.D:
            suit_str = "D"
        elif self.suit == Suit.S:
            suit_str = "S"
        elif self.suit == Suit.H:
            suit_str = "H"
        return suit_str

    def get_value(self):
        if self.value == 1:
            value_str = "A"
        elif self.value == 10:
            value_str = "T"
        elif self.value == 11:
            value_str = "J"
        elif self.value == 12:
            value_str = "Q"
        elif self.value == 13:
            value_str = "K"
        else:
            value_str = str(self.value)
        return value_str

    def __repr__(self):
        suit_str = self.get_suit()
        value_str = self.get_value()
        cardstr = f"{value_str}{suit_str}"
        if not self.faceup:
            cardstr = cardstr.lower()
        return f"Card {cardstr}"

print(Card("AC"))
```
